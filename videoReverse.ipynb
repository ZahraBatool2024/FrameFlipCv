```python
import cv2
cap = cv2.VideoCapture("car.mp4")
l=[]
c=1
while cap.isOpened():
    r,f=cap.read()
    if r == True:
        resized_frame = cv2.resize(f, (320, 240))
        filename= r"C:\Users\PMY\Videos\python zahra\reverse video\images\demo"+ str(c)+ ".jpg"
        l.append(filename)
        cv2.imwrite(filename, f)
        c=c+1
        cv2.imshow("videoreverse", f)
        if cv2.waitKey(25)&0xff ==ord("q"):
            break
    else:
        break

cv2.destroyAllWindows()

```

```python
l.reverse()
for i in l:
    img_new=cv2.imread(i)
    cv2.imshow("reversenew", img_new)
    if cv2.waitKey(25)&0xff==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
